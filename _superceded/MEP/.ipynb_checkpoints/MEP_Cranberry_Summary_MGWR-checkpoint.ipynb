```python
# read the subwatershed attributes table
filename = 'df_MEP_Monitoring_Avg_Geo_Merge.csv'
_ = pd.read_csv(os.path.join(odr,filename))
#_ = _[_.columns.drop(list(_.filter(regex='DEP2')))]
#_ = _[_.columns.drop(list(_.filter(regex='SLOPE_pct')))]
_.replace([np.inf, -np.inf], np.nan, inplace=True)
_.fillna(0,inplace=True)
print(_.info())
display(_.head())
df = _
del _ # clear temporary object from memory
```

```python
df['Cranberry_In_Terminus'] = df['LE5_CRANBERRY_1']>0
print(df['Cranberry_In_Terminus'])
```

```python
#def fn_HLR_Cran(df):
df['Cranberry_In_Terminus'] = df['LE5_CRANBERRY_1']>0
```

```python
# Calculate Metrics
df['NO3_GT_0p5'] = df['NOx']>0.5 # cranberry area (m2) in terminus zone
df['LE5_Cran_Area'] = df['LE5_CRANBERRY_1']*df['LE5_Shape_Area'] # cranberry area (m2) in terminus zone
df['Cran_HLR']=df['Qmeas']/df['LE5_Cran_Area'] # hydraulic loading rate (m d-1) to cranberry in terminus zone
df['Cran_Q_atHLR0p1']=df['Qmeas']/(df['LE5_Cran_Area']) # Q that cranberry could recieve at HLR = 0.1 m d-1

# calculate subwatershed NO3-N load g N d-1
df['L_NO3_sub'] = df['Qmeas']*df['NOx'] # g NO3-N day-1
# watershed areal N loading rate
df['LR_NO3_sub'] = df['L_NO3_sub'] /df['Shape_Area'] # g NO3-N m-2 day-1

# areal N loading rate to cranberry bogs
df['LR_NO3_cran'] = np.zeros
df['LR_NO3_cran'] = df['L_NO3_sub']/df['LE5_CRANBERRY_1']
df.to_csv(os.path.join(odr,'MEP_Cranberry_Summary.csv'))
```

```python
selected = ['Region_MEP','TN','Atten','NO3_GT_0p5','NOx','Qmeas','Cran_HLR','Cranberry_In_Terminus','LE5_Cran_Area','L_NO3_sub','LR_NO3_sub','LR_NO3_cran','Use_ResComMix','NaturalCover','LE5_NaturalCover']
_ = df[df['Cranberry_In_Terminus']==True][selected]
_.to_csv(os.path.join(odr,'MEP_Cranberry_Summary_Select.csv'))
```

```python
#import libpysal as ps
from mgwr.gwr import GWR
from mgwr.gwr import MGWR
from mgwr.sel_bw import Sel_BW
#df
# coords list of points in (x, y) format
coords = list(zip(df['Lon'], df['Lat']))
y = np.array(df['NOx']).reshape((-1,1))
x_names = ['HYDRCRATNG_No',
           'COVERNAME_Impervious',
           'SLOPE_D',
           'LE5_USEGENNAME_Commercial',
           'GT5_DEP2WATTBL_0',
           'GT5_USEGENNAME_Commercial']
n = len(x_names)
X = np.hstack([np.array(df[x]).reshape(-1,1) for x in x_names])
```

```python
# search for optimal bandwidth
bw = Sel_BW(coords, y, X).search(criterion='AICc')
# fit model with bandwidth
model = GWR(coords, y, X, bw, fixed=False, kernel='bisquare')
results = model.fit()
```
